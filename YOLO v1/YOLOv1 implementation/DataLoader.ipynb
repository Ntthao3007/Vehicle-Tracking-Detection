```python
from os import listdir
from PIL import Image
from torchvision import transforms
import torch
import cv2
import json
import random
```

```python
# Define transform which is applied to every single image to resize it and convert it into a tensor.
transform = transforms.Compose([
    transforms.Resize(448),
    #transforms.CenterCrop(448),
    transforms.ToTensor(),
    ])

# System paths to the train images and the label file.
# ToDo: Adapt this to only use the respective folder location.
train_files_path = 'C:/Users/alens/Desktop/Real-time-Object-Detection-for-Autonomous-Driving-using-Deep-Learning/YOLO v1/bdd100k/images/100k/val/'
target_files_path = 'C:/Users/alens/Desktop/Real-time-Object-Detection-for-Autonomous-Driving-using-Deep-Learning/YOLO v1/bdd100k_labels_release/bdd100k/labels/bdd100k_labels_images_val.json'

# Reference list to all the label categories for object detection.
category_list = ["traffic sign", "car", "rider", "motor", "traffic light", "person", "truck", "bus", "bike", "train"]


######################################
#    Load the images and the json    #
######################################
# All image names from the directory are loaded into the list train_files.
train_files = listdir(train_files_path)
# The json file containing the labels is loaded into the variable target_files.
f = open(target_files_path)
target_files = json.load(f)


kurcina = [] # Delete this (used for debugging).
# Here we store all training image tensors and training label tensors.
train_data_list = []
train_labels_list = []


# Loop over all image names from the directory.
for i in range(len(train_files)):
    # Delete this if-statemant (used for debugging).
    if i == 1:
        break
    
    ############################
    #    Extract the images    #
    ############################
    # Choose a random file from the directory and load it into the train_data_list.
    f = random.choice(train_files)
    train_files.remove(f)
    img = Image.open(train_files_path + f)
    img_tensor = transform(img) # Apply the transform to the image.
    train_data_list.append(img_tensor)
    
    kurcina.append(train_files_path + f) # Delete this
    
    ############################
    #    Extract the labels    #
    ############################
    # Find the json fragment with the same name as the randomly chosen image from above.
    for json in target_files:
        if json['name'] == f:
            img_label = json
            break

    # Loop over all labels of the respective json fragment. It can contain an arbitrary
    # amount of objects.
    for labels in range(len(img_label["labels"])):
        
        # Store the category index if its contained within the category_list.
        category = img_label["labels"][labels]["category"]         
        if category not in category_list:
            continue
        ctg_idx = category_list.index(category)
        
        # Store the bounding box information and rescale it by the resize factor.
        x1 = img_label["labels"][labels]["box2d"]["x1"] * (448/img.size[0])
        y1 = img_label["labels"][labels]["box2d"]["y1"] * (448/img.size[1])
        x2 = img_label["labels"][labels]["box2d"]["x2"] * (448/img.size[0])
        y2 = img_label["labels"][labels]["box2d"]["y2"] * (448/img.size[1])
        
        train_labels_list.append([ctg_idx,x1,y1,x2,y2])
        # Fill the new information inside the target tensor.
        target_tensor = torch.zeros(15, 7, 7)
        x_mid = abs(x2 - x1) / 2
        y_mid = abs(y2 - y1) / 2
        width = abs(x2 - x1)
        height = abs(y2 - y1)
        
        cell_dim = 448 / 7
        
        cell_pos_x = int(x_mid // cell_dim)
        cell_pos_y = int(y_mid // cell_dim)
        
        target_tensor[0][cell_pos_y][cell_pos_x] = 1
        target_tensor[1][cell_pos_y][cell_pos_x] = x_mid
        target_tensor[2][cell_pos_y][cell_pos_x] = y_mid
        target_tensor[3][cell_pos_y][cell_pos_x] = width
        target_tensor[4][cell_pos_y][cell_pos_x] = height
        #print(img.size)
        

# This code produces the bounding boxes for debugging.
%matplotlib inline        
color = (0, 255, 0)
thickness = 2
image = cv2.imread(kurcina[0])
image = cv2.resize(image, (448, 448),interpolation = cv2.INTER_AREA)
for i in range(len(train_labels_list)):
    start_point = (int(train_labels_list[i][1]),int(train_labels_list[i][2]))
    end_point = (int(train_labels_list[i][3]),int(train_labels_list[i][4]))

    cv2.rectangle(image, start_point, end_point, color, thickness)
    start_point = (int(train_labels_list[i][1]),int(train_labels_list[i][2])-10)
    cv2.putText(image, str(category_list[train_labels_list[i][0]]), start_point, cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
cv2.imshow('a', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
